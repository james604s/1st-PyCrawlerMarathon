## 作業目標

根據範例 ，完成以下問題：

* ① 取出今天所有的發文
* ② 如果想要依照類別分類，怎麼儲存會比較好？
* ③ 哪一個類別的文章最多




```python
import requests
from bs4 import BeautifulSoup
import re
import datetime
import json

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text, "lxml")

#分類類別與對應
classify_dict=dict()
for item in soup.find("div",class_="part_menu_2").find_all('li'):
    if item.string != "|":
        #類別名稱、編號對應
        k = re.split('[-|.]', item.find('a').get("href"))[-2] 
        v = item.string
        classify_dict.update({k:v})

#刪除總覽
classify_dict.pop('0')

#分類儲存 {key:[]}
today = datetime.date.today()
classify_title_dict = dict()
for k, v in classify_dict.items():
    classify_title_dict.update({v:list()})
    # for d in soup.find("div",class_="part_list_2").find_all('h3'):
    url = "https://www.ettoday.net/news/news-list-2021-07-21-"+str(k)+".htm"
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "lxml")
    soup = soup.find("div", class_="part_list_2")
    for item in soup.find_all("h3"):
        d = datetime.datetime.strptime(item.find(class_="date").text, "%Y/%m/%d %H:%M").date()
        if d == today:
            classify_title_dict[v].append(item.find_all('a')[-1].text)

#分類數量計算
classify_nums = dict()
for k, v in classify_title_dict.items():
    classify_nums.update({k:len(v)})
```

```python
with open('output.json', 'w') as f:
    result = json.dumps(classify_title_dict,indent=4, sort_keys=True, ensure_ascii=False)
    f.write(result)
```
